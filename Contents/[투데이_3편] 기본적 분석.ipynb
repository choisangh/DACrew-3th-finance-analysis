```python
def low_rank(data_frame, col):
  result = data_frame.sort_values(by = col)
  return result, result[col].rank()

def high_rank(data_frame, col):
  result = data_frame.sort_values(by = col, ascending=False)
  return result, result[col].rank(ascending=False)
```

```python
# í˜„ì¬ ì €í‰ê°€ ëœ ìƒìœ„ 5ê°œ ê¸°ì—… ì°¾ê¸° 
def evaluation(data_frame,n=5):

  df_rank = pd.DataFrame()
  # ë‚®ì„ìˆ˜ë¡ ì €í‰ê°€ ëœ ì§€í‘œ
  low = ['PER','PBR']
  for l_col in low:
    df_rank[f'{l_col}_ë­í‚¹'] = low_rank(data_frame, l_col)[1]

  # ë†’ì„ìˆ˜ë¡ ì €í‰ê°€ ëœ ì§€í‘œ
  high = ['EPS','ROE']
  for h_col in high:
    df_rank[f'{h_col}_ë­í‚¹'] = high_rank(data_frame, h_col)[1]

  # ê° ì§€í‘œì— ê°€ì¤‘ì¹˜ë¥¼ ì¡°ì ˆí•˜ì—¬ ì¢…í•©ìˆœìœ„ ê²°ì •
  df_rank['ì¢…í•©ìˆœìœ„'] = df_rank['PER_ë­í‚¹']*0.3 + df_rank['PBR_ë­í‚¹']*0.1 + df_rank['EPS_ë­í‚¹']*0.1 + df_rank['ROE_ë­í‚¹']*0.5 
  df_rank = df_rank.rank().sort_values(by='ì¢…í•©ìˆœìœ„')

  ticker = [stock.get_market_ticker_name(ticker=i) for i in df_rank.index]
  df_rank['ì¢…ëª©'] = ticker
  
  return df_rank.head(n)
```
