#### **당기순이익**
일정기간(당기)동안 매출액에서 순비용항목을 차감한 기업의 순수 이익입니다.<br>
(순비용항목이란 매출원가, 판매 및 관리비, 기타비용, 금융비용, 법인세비용을 차감한 것)<br>
> 당기순이익 = 매출액 - 매출원가 - 판매관리비 - 기타비용 - 금융비용 - 법인세비용


|(-) / (+)|구성|예시
|--|--|--
|    |**매출액**|판매량 x 판매단가
|(-)|매출원가|원재료, 전력비용, 감가상각비
|(-)|판매 및 관리비|인건비, 광고비
|    |**=   영업이익**|
|(+) (-)|기타손익|기타수익, 기타비용<br>임대료, 외환차익차손, 재고자산폐기손실
|(+) (-)|금융손익|금융수익, 금융비용<br>이자수익
|    |**=   법인세 차감전 순이익**|
|(-)|법인세|
|    |**=   당기순이익**|

<br>

여기서 중요한 것은 영업이익과 당기순이익의 구분입니다!<br>
- **영업이익**은 매출액에서 매출원가와 판매관리비만을 제외한 이익입니다. 따라서 순수 기업활동을 통해서 거두어 들인 이익으로 볼 수 있습니다.
<br>
- **당기순이익**은 영업이익에서 조달한 자금 내역 등을 제외한 투자자와 주주에게 돌아가는 수익입니다.

그럼 **기업의 핵심능력**을 파악하고 싶다면 둘 중 어느것이 더 적합할까요?<br>



정답은 영업이익입니다!<br>
당기순이익은 영업 외 손익도 포함되어 있기  때문에 기업 본연의 사업 분야 성과를 파악하기에 더 적합한 지표입니다.

#### **주당순이익(EPS: Earning Per Share)**

말 그대로 1주당 기업의 당기순이익입니다.<br>
다시말해, 보통주 1주당 분배될 수 있는 이익의 크기입니다.<br> 
이 지표는 기업 전체의 경영 성과에 기반하여 주식 한 주당 얼마의 순이익이 발생했는지 확인할 수 있습니다.
> EPS = 당기순이익 / 총 발행 주식수 <br>

따라서 EPS가 낮다는 것은 기업이 이익을 충분히 내지 못하고 있는 것이겠죠?
그 원인으로는 아래와 같은 이유가 있습니다.

- 기업의 비효율적 경영
- 부채비율이 높아 재무부담 증가
- 연구개발비에 지나친 투자
- 지나치게 보수적인 회계원칙 채택
- 기업의 수익성 하락 또는 유상증자로 인해 총 발행주식수 증가

#### **주가수익비율(PER: Price Earnings Ratio)**

주당순이익(EPS) 대비 주가를 측정하는 지표로 투자금을 회수하는데 걸리는 시간입니다.

> PER = 1주당 주가 / 주당순이익(EPS) = 시가총액 / 당기순이익


종목별로 조금의 차이가 있지만 보통 10 이하이면 저평가되어있다고 판단할 수 있습니다. <br>
(시가총액이란? 주가에 총 주식 수를 곱한 지표입니다.)

#### 자산
자산은 유동자산과 비유동자산으로 구성됩니다.<br>
1~2년 이내에 현금화가 가능하면 **유동자산**, 그렇지 않으면 **비유동자산**으로 구분됩니다.<br>
- 유동자산 
    - 재고자산 : 원재료, 재공품, 제품, 상품으로 곧 팔릴 것으로 보기 때문에 유동자산에 속합니다.<br>

    여기서 중요한 점은 재고자산의 비율이 작년에 비해 과도하게 상승했다면 좋은 신호가 아닙니다.<br>

    - 매출채권 : 쉽게 외상으로 생각하면 됩니다. 물건을 팔고 대가로 현금이 아닌 채권으로 받은 것 입니다.
    - 현금 및 현금성자산 : 현금성자산은 3개월 이내에 현금으로 받을 수 있는 자산을 의미하며 만기가 짧은 금융상품은 현금에 준하는 것으로 보아 유동자산에 속합니다.<br>

     단기금융상품이 현금 및 현금성자산 바로 아래 적혀있는데 만기가 1년 이내인 정기예금 또는 정기적금 등이 해당됩니다. 
    - 기타유동자산(투자자산)

- 비유동자산 
    - 유형자산 : 공장, 토지, 건물, 기계장치 등이 해당됩니다.
    - 무형자산 : 특허, 산업재산권 등이 해당됩니다.
    - 기타 금융자산

# 2. 기본적 분석 이해
## 기본적 분석이란?

기업의 내재적 가치를 파악하여 시장내에 기업의 위치를 예측하고, 대략적인 매매 시기를 판단하는 분석 방법입니다.<br>

성공한 주식 투자가로 자주 언급되는 워렌 버핏, 피터 린치가 기본적 분석을 사용한 투자가입니다!<br>
주가가 내재 가치보다 낮다고 판단될 때 매수하고, 주가가 내재 가치보다 높다고 판단될 때 매도하는 방식으로 투자를 하는 것이 기본적 분석의 투자 방법입니다.

## 2-1. 재무비율분석
앞서 설명한 재무제표에서 2개 이상의 항목 비율을 분석해 투자하는 방법이 있습니다.<br>
그 중 재무상태표와 손익계산서의 지표를 활용하는 것이 **재무비율분석**입니다.<br>
재무비율은 유동성지표, 안정성지표, 수익성지표, 활동성지표, 부상비율, 이익지표로 총6가지 종류의 지표가 있습니다.

#### 유동성지표
단기부채를 상환할 수 있는지 능력을 나타내는 지표입니다.
* 유동비율(CR : Current Ratio)
> 유동비율 = 유동자산/유동부채
  
  - 단기부채를 상환하기 위해 현금화할 수 있는 능력으로 기업이 보유하는 신용능력을 판단할 수 있습니다.
  - 높을수록 단기부채를 쉽게 상환할 수 있음을 의미합니다.
  - 통상적으로 200% 이상인 기업은 건전한 것으로 인식됩니다.


* 당좌비율(QR : Quick Ratio)
> 당좌비율 = 당좌자산 / 유동부채 =  (유동자산 - 재고자산 - 선급금) / 유동부채

  - 당좌자산 : 현금, 예금, 유가증권, 매출채권 등
  - 단기부채를 부담할 수 있는 기업의 능력을 측정하는 지표로 기업의 현금보유 상태를 가장 잘 알려줍니다.
  - 통상적으로 100% 이상이면 좋은 것으로 판단됩니다.

* 현금비율(CAR : Cash Ratio)
> 현금비율 = (현금 + 시장성 유가증권)/유동부채

  - 단기부채의 상환능력을 측정하는 지표 중 가장 엄격하며, 기업의 현금보유 상태를 가장 잘 알려주는 지표입니다.
  - 현금비율이 너무 높을 경우 보수적인 경영정책을 암시할 수 있으며, 너무 낮은 경우는 현금부족 사태에 직면할 수 있어 적절한 수치를 유지해야 합니다.

```python
liquidity = is_bs.copy()
liquidity = liquidity[(liquidity['구분']=='유동자산')|(liquidity['구분']=='재고자산')|
                      (liquidity['구분']=='선급비용')|(liquidity['구분']=='유동부채')|
                      (liquidity['구분']=='단기금융상품')|(liquidity['구분']=='단기당기손익-공정가치금융자산')|
                      (liquidity['구분']=='당기손익-공정가치금융자산')|(liquidity['구분']=='현금및현금성자산')]
liquidity.set_index('구분', inplace=True)
liquidity
```

```python
for col in liquidity.columns:
  liquidity[col] = liquidity[col].astype('int64')
  liquidity[col] = liquidity[col] #/100000000

liquidity = liquidity.T
liquidity['<유동비율>']=round(liquidity['유동자산']/liquidity['유동부채']*100,2)
liquidity['<당좌비율>']=round((liquidity['유동자산']-liquidity['재고자산']-liquidity['선급비용'])/liquidity['유동부채']*100,2)
liquidity['<현금비율>']=round((liquidity['현금및현금성자산']+liquidity['단기금융상품']+
                           liquidity['단기당기손익-공정가치금융자산']+liquidity['당기손익-공정가치금융자산'])/liquidity['유동부채'] * 100,2)
liquidity[liquidity.columns[-3:].to_list()+liquidity.columns[:-3].to_list()]
```

#### 안정성지표
기업의 중장기적 채무이행능력을 나타내는 지표입니다.
* 부채비율(DR: Debt Ratio)
> 부채비율 = 총부채 / 총자산(총자본)
  
  - 총자본 중 채권자들이 제공한 자금의 비율로 재무레버지리의 크기를 나타냅니다.
  - 낮은 부채비율은 차입금을 상환하기 위한 자산의 비율이 높은 것이기 때문에 채권자 및 주주들이 선호합니다.
  - 일반적으로 50% 보다 낮으면 안정적인 것으로 판단됩니다.


* 부채-자기자본비율(DER: Debt-equity Ratio)
> 부채-자기자본비율 = 총부채/자기자본

  - 자기자본보다 부채가 얼마나 더 많은지에 대한 비율입니다.
  - 부채는 채권으로 이자비용이 발생하고, 자기자본은 주식으로 이자비용은 없으나 성과에 대한 배당을 합니다.
  - 일반적으로 100%보다 높을수록 기업의 위험과 요구수익률이 높고, 낮을수록 위험이 낮고 안정적으로 판단됩니다.

#### 수익성지표
기업의 이익창출능력을 나타내는 지표입니다.
* 매출액영업이익률(OPM: Operating Profit Margin)
> 매출액영업이익률 = 영업이익/순매출액
  
  - 매출액 한 단위가 창출하는 영업이익입니다.
  - 일반적으로 높을수록 효율적인 영업을 의미하나, 지나치게 높은 경우 충분한 투자가 되지 않고 있음을 예상할 수 있습니다.

* 총자산이익률(ROA: Return on Assets)
> 총자산이익률 = 당기순이익/총자산(자본+부채) = 순이익/순매출액 x 순매출액/총자산

  - 기업이 보유하고 있는 자산을 얼마나 효율적으로 이용하는지를 나타내는 지표입니다. 즉, 총 자산으로 얼마나 많은 순수익을 올렸는지 알 수 있습니다.
  - 비율이 높을수록 수익성이 양호하며, 효율적인 영업을 한다고 판단할 수 있습니다.
  - 지나치게 높은 경우 R&D 투자가 미흡함을 예상할 수 있습니다.

* 자기자본이익률(ROE: Return on Equity)
> 자기자본이익률 = 당기순이익/자기자본 = ROA/자기자본비율 = ROA/(1 - 총부채/총자산)

  - 주주들의 투자자본인 자기자본을 얼마나 효율적으로 활용했는지 측정하는 지표입니다.
  - 재무분석에서 가장 중요한 지표로 높을수록 주주들의 이익이 많아진다는 의미입니다.
  - ROA나 부채비율이 높을수록 상승하므로, 어떤 요인이 ROE를 상승시키는지 확인해야 합니다.

```python
profit = is_bs.copy()
profit = profit[(profit['구분']=='수익(매출액)')|(profit['구분']=='영업이익')|
                (profit['구분']=='당기순이익(손실)')|(profit['구분']=='자본총계')|
                (profit['구분']=='부채총계')]

profit.set_index('구분', inplace=True)
```

```python
for col in profit.columns:
  profit[col] = profit[col].astype('int64')
  profit[col] = profit[col]/100000000
profit = profit.T

profit['<매출액영업이익률>'] = round(profit['영업이익']/profit['수익(매출액)'] * 100,2)
profit['<총자산이익률(ROA)>'] = round(profit['당기순이익(손실)']/profit['자본총계'] * 100,2)
profit['<자기자본이익률(ROE)>'] = round(profit['<총자산이익률(ROA)>']/(1-(profit['부채총계']/profit['자본총계'])),2)

profit[profit.columns[-3:].to_list() + profit.columns[:-3].to_list()]
```

## 2-2. 레버리지 분석

레버리지는 지렛대(lever)에서 파생된 말입니다.
지렛대를 이용하면 적은 힘으로도 무거운 물건을 들어올릴 수 있죠!?<br>
중심점으로부터 먼 곳에서 작동할 수록 더 적은 힘으로 같은 무게의 물건을 들어올릴 수 있습니다.

기업의 금융활동에 지렛대의 원리를 적용하면 어떨까요?
적은 자본으로 수익률을 극대화시킬 수 있는 방법은 무엇일까요?


정답은 '**부채**'입니다!<br>

기업은 '빚'을 내어 부족한 돈을 빌려 투자하는 방법을 통해 적은 자본으로도 높은 수익률을 달성할 수 있습니다.<br>
또한 중심점인 '고정비'를 움직여 자본의 이익률을 높일 수 있습니다.<br>
레버리지의 종류에는 영업레버리지, 재무레버리지, 결합레버리지가 있습니다.

### 영업 레버리지 (Operation Leverage), DOL

기업의 영업비용 중, **고정비를 부담하는 정도**를 의미합니다.<br>
고정 영업비의 존재로 영업이익이 커지는 현상이며, 고정비를 부담하지 않는 기업에서는 영업레버리지 효과가 발생하지 않습니다.

>  DOL = 영업이익 변화율 / 매출액 변화율 = (매출액 - 변동비) / (매출액-변동비-고정비)



### 재무 레버리지 (Financial Leverage), DFL
기업이 조달한 총자본 중, **타인자본(이자비용)이 차지하는 비율**을 의미합니다.<br>
타인자본 의존도가 높을수록 크고, 영업이익이 커질수록 낮아집니다. 

>  DFL =  순이익 변화율/ 영업이익 변화율

### 결합 레버리지 (Combined Leverage), DCL
영업레버리지와 재무레버리지를 결합한 레버리지입니다. <br>
영업고정비와 재무고정비 중 하나라도 존재하면 1보다 큰 값을 가집니다.
고정비가 많거나 타인자본 의존도가 높은 기업일수록 결합 레버리지가 높게 나타납니다.

> DCL = 순이익 변화율 / 매출액 변화율 =  영업레버리지(DOL) x 재무레버리지(DFL)

## 2-3. 상대가치평가모형(주가배수모형)

### 주가수익비율(PER)

주가를 주당순이익으로 나눈 것으로, 주가가 수익의 몇 배인지를 알 수 있는 지표입니다.

> PER = 1주당 주가 / 주당순이익(EPS) = 시가총액 / 순이익

- 기업의 수익 창출 능력을 가장 잘 반영한 지표입니다.
- 수치가 낮을수록 수익에 비해 주가가 낮기 때문에 기업의 가치에 비해 주가가 저평가 되어있다는 의미로 해석할 수 있습니다.
- 통상적으로 수치가 높을수록 변동성이 커지며 투자위험이 높아집니다.

### 주가순자산비율(PBR)

주가를 주당순자산가치로 나눈 것으로, 기업의 순자산에 비해 1주당 몇 배가 거래되는지를 알 수 있는 지표입니다.

> PBR = 시장가치(주가)/주당순자산 = 순이익/매출액 x 매출액/총자산 x 총자산/자기자본 x PER =  ROE x PER

- PBR 이 1보다 작으면 기업의 가치에 비해 주가가 낮다고 해석할 수 있습니다.
- 통상적으로 PBR 이 낮은 기업이 그렇지 않은 기업에 비해 높은 주가수익률을 보입니다.
- 가치투자자들은 PER과 함께 이 지표를 사용하여 높은 수익률을 낸다고 합니다.

### EV/EBITDA 비율

기업가치가 영업활동을 통해 얼마만큼의 이익을 벌었는가를 나타내는 지표입니다.

> EV = 시가총액 + 순차입금(이자지급성 부채 - 현금 및 유가증권) = 주주가치 + 채권자가치<br>
> EBITDA = 영업이익 + 감가상각비 + 무형자산상각비

- 당기순이익을 기준으로 평가하는 PER의 한계점을 보완해주는 지표입니다.
- 비율이 낮을수록 기업 가치에 비해 주가가 저평가 되었다고 해석할 수 있습니다.

# 3. 기본적 분석 활용 투자전략 구현

기업 가치에 비해 주가가 저평가 되어있는 종목을 찾는 것이 많은 투자자들의 성공 비법입니다.<br>
앞서 공부한 지표를 이용하여 저평가 되어있는 종목을 찾아봅시다!

```python
#!pip install pykrx
#!pip install finance_datareader
import FinanceDataReader as fdr
from pykrx import stock
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# KRX NASDAQ, NYSE, AMEX, SP500
krx_list = fdr.StockListing('KOSPI')
print(krx_list.shape)
krx_list.head()
```

업종별로 PER 수치에는 편차가 존재하므로 함께 비교하는 것은 무의미합니다.<br>
따라서 업종에 별로 데이터를 생성해줍니다.

```python
# 업종별로나누기
sector_list = krx_list['Sector'].unique()

# 업종별 종목코드 저장
건설철강금속 = []
금융및보험 = []
소매업 = []
서비스업 = []
for idx, row in krx_list.iterrows():
  if "건설" in str(row['Sector']) or ("철강" in str(row['Sector'])) or ('금속' in str(row['Sector'])):
    건설철강금속.append(row['Symbol'])
  elif ("금융" in str(row['Sector'])) or ('보험' in str(row['Sector'])):
    금융및보험.append(row['Symbol'])
  elif "소매" in str(row['Sector']):
    소매업.append(row['Symbol'])
  elif "서비스" in str(row['Sector']):
    서비스업.append(row['Symbol'])
```

```python
# BPS	PER	PBR	EPS	DIV	DPS 지표 가져오기
st = stock.get_market_fundamental("20211230", market="KOSPI")
st.head(3)
```

```python
# 업종별 데이터 저장하기
df_건설철강금속 = st.loc[건설철강금속]
df_금융및보험 = st.loc[금융및보험]
df_소매업 = st.loc[소매업]
df_서비스업 = st.loc[서비스업]
df_소매업.head()
```

```python
# 전처리 함수
def preprocess(data_frame):

  # 안전한 종목 선정을 위해 순이익이 없는 종목은 제거
  data_frame = data_frame.replace(0,np.NaN)
  result = data_frame.dropna()

  # ROE 지표 추가
  result['ROE'] = result['EPS']/result['BPS'] * 100
  
  # 종목이름 추가
  ticker = [stock.get_market_ticker_name(ticker=i) for i in result.index]
  result['종목'] = ticker

  print(f'전처리 전:{data_frame.shape} -> 전처리 후: {result.shape}')

  return result

df_건설철강금속 = preprocess(df_건설철강금속)
df_금융및보험 = preprocess(df_금융및보험)
df_소매업 = preprocess(df_소매업)
df_서비스업 = preprocess(df_서비스업)
```

```python
df_서비스업.head()
```

```python
def low_rank(data_frame, col):
  result = data_frame.sort_values(by = col)
  return result, result[col].rank()

def high_rank(data_frame, col):
  result = data_frame.sort_values(by = col, ascending=False)
  return result, result[col].rank(ascending=False)
```

```python
# 현재 저평가 된 상위 5개 기업 찾기 
def evaluation(data_frame,n=5):

  df_rank = pd.DataFrame()
  # 낮을수록 저평가 된 지표
  low = ['PER','PBR']
  for l_col in low:
    df_rank[f'{l_col}_랭킹'] = low_rank(data_frame, l_col)[1]

  # 높을수록 저평가 된 지표
  high = ['EPS','ROE']
  for h_col in high:
    df_rank[f'{h_col}_랭킹'] = high_rank(data_frame, h_col)[1]

  # 각 지표에 가중치를 조절하여 종합순위 결정
  df_rank['종합순위'] = df_rank['PER_랭킹']*0.3 + df_rank['PBR_랭킹']*0.1 + df_rank['EPS_랭킹']*0.1 + df_rank['ROE_랭킹']*0.5 
  df_rank = df_rank.rank().sort_values(by='종합순위')

  ticker = [stock.get_market_ticker_name(ticker=i) for i in df_rank.index]
  df_rank['종목'] = ticker
  
  return df_rank.head(n)
```

```python
df_건설철강금속_5 = evaluation(df_건설철강금속)
df_금융및보험_5 = evaluation(df_금융및보험)
df_소매업_5 = evaluation(df_소매업)
df_서비스업_5 = evaluation(df_서비스업)
df_서비스업_5
```
