# **INTRO**
본격적인 콘텐츠 업로드에 앞서, 금융데이터 분석에 대한 이해를 돕기 위해 금융시장에 대해서 간단하게 알아보도록 하겠습니다.


## **1. 금융시장이란?**
금융시장이란, 간단하게 말하자면 자금 수요자와 공급자 사이에 <b>금융거래가 이루어지는 장소</b>입니다.  

즉, 자금에 여유가 있는 공급자가 자금이 필요한 수요자에게 자금을 제공하고, 수요자는 공급자에게 이자, 배당금과 같은 일종의 <b>benefit</b>을 제공 합니다.

금융시장은 이러한 수요자, 공급자 간의 **중개역할**을 해줍니다.

<br>  

| 거래상품 성격에 따른 분류 ||
| :--: | :-- |
| 증권시장 | 유가증권(주로 주식 및 채권)이 거래되는 시장 |
| 단기금융시장(화폐시장)| 콜,CP 등 만기 1년 미만의 단기 자금조달수단이 거래되는 시장|
| 예금·대출시장|예금상품 및 대출상품이 거래되는 시장|
| 외환시장 |서로 다른 두 가지 통화를 교환하는 시장|
| 파생상품시장 |금리·주가·환율 변동의 위험 회피를 위해 형성된 시장|
<br>

금융시장은 취급하는 상품에 따라서 여러 시장으로 나뉩니다.

저희에게 익숙한 코스피, 코스닥과 같은 주식 시장은 증권 시장 중 하나입니다.

저희 콘텐츠는 **주식 투자**와 관련 데이터를 주로 활용할 예정이므로 <b>증권시장</b>에 대한 내용을 중점적으로 다룰 계획입니다.

## **2. 증권 분석이란?**



주식에 투자하기 위해서는 해당 기업에 대한 가치와 더불어 다양한 factor에 대한 분석이 이루어져야 합니다.<br>
쉽게 말해서, '요즘 경제 상황은 어떠한지', '현재 주가가 고점인지 저점인지', '미래에도 유망한 산업인지'와 같은 증권 분석의 과정이 필요합니다.
<br><br>

<br>

### **증권의 기본적 분석(Fundamental anlysis)**
기본적 분석은 내재적 가치를 평가하는 투자 분석 기법입니다. 재무제표를 통해서 기업의 성장세와 안정성을 판단하고 경제분석, 산업분석을 통해 장기적인 주가의 흐름을 예측 합니다.
* 경제 분석 : 현재 경제 상황은 어떤가? **(2편에서 다룰 예정)**
> 경제성장률, 물가, 통화량, 금리, 환율 등 <br>
* 산업 분석 : 미래에도 유망한 산업인가? **(3편에서 다룰 예정)**
> 산업성장률, 시장규모, 경쟁구조, 비용구조 등 <br>
* 기업 분석 : 투자할 가치가 있는 기업인가? **(3편에서 다룰 예정)**
> 기업영업실적, 재무구조, 제품시장성, 기업이미지 등  <br>
<br>

### **증권의 기술적 분석(Technical anlysis)**
기술적 분석은 오직 주식의 변동가격 움직임 자체만을 다루는 투자 분석 기법입니다. 기술적 분석은 투자자의 심리적인 요인을 반영하여 기본적 분석의 한계점을 보완하는 장점을 가지고 있습니다.
흔히 말하는 '차트를 본다'라는 표현은 기술적 분석을 의미합니다.<br><br>

<img src="https://marginfan.com/files/attach/images/147/092/008/0a5591cdbc33a988805dede6cc63c934.png" width="500px"><br><br>


좀 더 직관적으로 표현하자면, 과거 비트코인 커뮤니티에서 유행했던 '기영이 차트'도 기술적 분석을 희화화한 밈이라고 할 수 있습니다.

* 기술적분석 : 현재 주식 차트의 추세가 어떠한가? **(4편에서 다룰 예정)**
> 다우이론, 엘리어트 파동이론, RSI, MACD, OBV등 기술지표

<br>

앞서 설명한 이론을 바탕으로 실제 금융 데이터를 통해 증권 분석하는 일련의 과정을 콘텐츠로 담을 예정이니 많은 기대와 관심 부탁드립니다🔥🔥🔥
