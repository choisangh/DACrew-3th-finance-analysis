# 1. 기술적 분석의 이해

## 기술적 분석이란?


기본적 분석은 주식의 가격이 고평가되었는지, 저평가되었는지를 결정하기 위해 주식의 내재가치를 파악하려는 시도라고 볼수있습니다.<br>
반면에 **'기술적 분석'**은 주식시장의 **과거가격의 패턴**을 분석하여 미래가격의 움직임이나 시장의 경향을 예측하는 것입니다.<br>

기본적 분석은 시장가격이 언젠가는 내재가치에 수렴한다는 가정 하에 주식의 실현가치에 접근하려고 노력하는 반면에<br>
기술적 분석또한 역시 시장가격을 보긴 하지만 이는 주식의 가격움직임을 예측하려는 시도이고 가격이 움직이는 이유에 대해서는 관심을 가지지 않습니다.<br>

기술적 분석가는 과거의 가격이나 거래량 자료의 조사에 기초하여 미래의 경향을 예상하고 주가의 움직임을 예측합니다.<br>
다시 말해 기술적 분석은 **과거성과를 이용하여 미래성과를 예측하는 것**이라 할 수 있습니다.<br>


### 기술적 분석의 기본가정
기술적 분석은 지금까지 이어져온 가격의 움직임을 바탕으로 미래가격을 예측합니다.<br>
기술적 분석이 의미가 있으려면 다음과 같은 가정이 현실에서도 충족되어야 합니다.<br>

1. 상품, 서비스, 주식가격 등의 시장가치는 단지 수요공급에의해 결정됩니다.
2. 수요와 공급은 공포와 욕심에 의해 합리적 행동과 비합리적 행동을 포함하는 많은 요소들에 의해 발생합니다.
3. 주가는 장기간 존속하면서 추세 속에서 움직입니다. 매우 짧은 기간의 출렁거림은 무시될 수 있습니다.
4. 수요와 공급의 변동요인을 결정하기는 때때로 어렵습니다. 반면에 실체적인 수요와 공급은 시장가격의 반응에 의해 조사될 수 있습니다.

결과적으로 기술적 분석은 수요와 공급의 충분한 효과가 시장가치(주가)에 반영되는 데는 시간이 걸리므로 주가움직임 패턴을 잘 분석하면 초과수익을 얻을 수 있다는 것입니다.

### 기술적 분석의 종류

1. 추세분석<br>
일반적으로 가격은 한번 어떠한 추세를 타기 시작하면 이 추세가 상당기간 동안 지속되는 경향이 있습니다. 이 같은 성질을 이용한 기술적 분석 방법을 추세분석이라 합니다.<br>

2. 패턴분석<br>
추세가 변화하면서 가격은 어떠한 패턴을 나타내는 경우가 있는데, 이를 분석하여 실제 가격의 움직임과 비교해 앞으로의 추세를 예측하는 분석 방법을 '패턴 분석'이라고 합니다.

3. 장세(시장강도)분석<br>
현재의 주가나 환율이 과매수권 혹은 과매도권 인지를 판단하기 위해 보조지표를 활용하여 매수와 매도 구간을 파악하는 분석 방법입니다. 주로 이동평균선, MACD, RSI, 스토캐스틱 등과 같은 기술적 지표를 이용하는 분석 방법입니다.<br>

4. 심리적 분석<br>
일반적인 분석 방법과 조금 다른 형태의 분석 방법으로, 시장 참여자들의 심리를 파악하여 그와 반대되는 포지션에 진입하고자 할 때 사용되는 분석 방법 입니다. 예를 들어 많은 투자자들이 매수 포지션을 보유하고 있을 경우 반대로 매도 포지션에 진입헤 다소 리스크가 큰 거래를 하는 투자자들이 주로 사용합니다.<br>

5. 시장구조분석<br>
시장의 움직임을 장시간 관찰하고 연구해 시장 변동원리를 파악하여 가격을 예측하는 분석 방법입니다. 대표적인 분석 방법으로 엘리어트 파동 이론이 있습니다.<br>


### 기술적 분석의 장점
1. 기술적 분석은 기본적 분석에 비해 용이하고 간단하여 한눈에 알아볼 수 있습니다.<br>

2. 증권의 매매시점을 포착하기 수월합니다.<br>

3. 기술적 분석은 주가에 영향을 미치는 **심리적인 요인**을 반영하므로 기본적 분석의 한계점을 보완할 수 있습니다.<br>

### 기술적 분석의 한계
1. 투자자들은 같은 차트를 보면서도 자기만의 분석 방법을 통하여 서로 다른 해석을 내놓게 됩니다. 이는 일반적으로 사람의 심리가 **자신에게 유리**하도록 해석하려는 경향이 있기 때문이며, 100% 객관적인 시선으로 시장을 바라보고 이를 분석하는 것은 쉽지 않다는 것입니다.<br>

2. 이처럼 실제 가격의 움직임은 시장 참가자들에게 달렸기 때문에 이를 예측하는 것은 불가능하다는 '랜덤워크 이론'과 같이 기술적 분석에 대해 회의적인 시선을 갖고 있는 사람들이 존재하는 것도 사실입니다.<br>

3. 하지만, 많은 성공한 투자자들이 기술적 분석을 매우 가치 있는 투자 도구라고 주장하고 있는 만큼 단순히 부정적인 시선 만을 가지고 이를 무시해서는 안될 것입니다.

# 2. 차트(Chart)

차트는 과거의 가격 변동을 그래프로 표현한 것으로 지금까지 축적된 패턴을 이용하여 다양한 정보를 읽을 수 있습니다. 그래프의 형태나 계산식 등을 이용하여 과거를 바탕으로 미래를 예측하는 **기술적 분석**을 가능하게 해줍니다.<br>

또한, 차트에는 시장 참가자들의 심리가 짙게 반영되어 있습니다. 투자자들은 기본적으로 똑같은 차트를 보고 거래를 하기 때문에 서로 비슷한 예측을 하고 포지션에 진입하게됩니다.<br>

예를 들어 주식이 곧 하락할 것이라는 신호가 차트에 나타나게 될 경우, 보유중인 포지션을 매도하고자 하는 투자자들이 다수 나타나게 됩니다. 그렇게 되면 실제로 매도 물량이 쏟아져 나와 주가가 하락하는 결과로 이어질 수 있습니다. 이는 바로 투자자들의 심리가 시장을 움직이는 결과를 초래한 것으로 볼 수 있습니다.<br>

이 같은 시장 참가자들의 심리를 재빨리 파악할 수 있다면, 이익을 얻을 수 있는 기회도 늘어나게 될 것입니다.


## 차트의 종류

주식의 차트는 거래된 가격과 거래량을 표시한 것으로 주가의 변동을 통해 예측을 하기 위한 보조 지표로 쓰인다고 합니다. 크게 봉(캔들)차트와 바차트, 그리고 라인차트까지 3가지의 종류가 있습니다.<br>


![이미지](https://user-images.githubusercontent.com/108919698/182627962-1fd191a8-ab09-4826-9d67-9600279b5b37.png)<br>
먼저 라인 차트는 종가 정보만으로 형성된 차트로, 주식 거래 분석을 위해 잘 쓰이는 차트는 아닙니다. 반면 캔들 차트와 바 차트는 국내와 해외에서 주식 거래 분석을 위해 흔히 쓰이는 차트로, 캔들 차트는 국내와 일본에서 주로 쓰이고 있고, 바차트는 미국에서 주로 쓰이고 있습니다.<br>

캔들(봉)차트와 바 차트는 모양과 색이 의미하는 바는 다르지만, 고가,저가,시가,종가 까지 4가지의 정보를 제공하고 있습니다. 국내 대다수의 증권사에서는 캔들 차트를 제공하고 있습니다.<br>



## 차트의 의미

그럼 주로 사용되는 봉 차트와 바 차트에서 각 요소들이 의미하는 것은 무엇인지 살펴 보겠습니다. 모양은 다르지만 봉 차트와 바 차트는 모두 동일한 정보를 제공하고 있습니다.<br>

![이미지](https://user-images.githubusercontent.com/108919698/182627948-01da7eb0-6857-4c0d-9e7d-0f674c606052.png)<br>

캔들 차트는 양봉과 음봉으로 이루어져 있습니다. 양봉과 음봉은 봉과 꼬리로 구성되어 있는데요. 꼬리의 가장 위는 그날 거래된 가격중 가장 높은 금액인 고가, 그리고 꼬리의 가장 아래는 그날 거래 가격 중 가장 낮은 금액인 저가를 의미합니다. 고가와 저가는 양봉과 음봉이 동일합니다. 봉은 양봉과 음봉이 서로 다른 요소를 의미하는데요. 먼저 양봉은 봉의 가장 윗부분이 그날 마지막 거래 가격인 종가를, 양봉의 가장 아랫부분은 그날의 첫 거래 가격인 시가를 의미합니다. 즉, 양봉은 처음 거래된 가격보다 가장 마지막으로 거래된 가격이 올랐을 때 빨간색 양봉으로 표시됩니다. 반대로 음봉은 봉의 가장 위가 시가, 봉의 가장 아래가 종가로, 그날 처음 거래된 가격보다 낮은 가격으로 마지막 거래가 이뤄졌을때 음봉, 즉 파란색으로 표시되는 것이죠.<br> 

![이미지](https://user-images.githubusercontent.com/108919698/182627954-a4bae489-4c1e-47ce-a7dd-ae440a604ed6.png)<br>

바 차트도 모양만 다를 뿐 의미하는 바는 동일합니다. 단, 주의하셔야 할 점이 봉 차트에서는 빨간색이 상승세의 주가이지만 바 차트에서는 빨간색은 하락세의 주가이고, 반대로 초록색이 상승세의 주가라는 점입니다. 그 외에는 바의 가장 윗부분이 고가, 가장 아랫부분이 저가이며, 오른쪽 왼쪽으로 튀어나온 '틱'이 각각 종가와 시가를 의미합니다. 바 차트에서는 종가와 시가를 나타내는 오른쪽 또는 왼쪽으로 튀어나온 바를 '틱'이라고 부릅니다.<br>
* 틱(Tick):틱은 1호가단위로, 최소의 가격 변동 단위를 의미합니다.

# 3. 차트 기술지표 이해

기술적지표는 흔히 **보조지표** 라고도 불리는데 다양한 각도와 계산식, 그리고 통계 등을 바탕으로 앞서 배운 기본적 분석 방법들과 조합하여 보다 폭 넓은 시장 예측을 가능하게 도와주는 차트 분석 도구를 말합니다. 기술적 지표는 주로 적절한 매수시점과 매도 시점을 파악하기 위해 사용되며, 감정의 기복을 배제한 객관적인 매매의 기준을 세우는데 활용됩니다.<br>


## 기술적 지표의 종류

* 먼저, 기술적 지표는 성격에  따라 크게 5가지로 구분됩니다.<br>

1. 추세지표<br>
추세의 진행 방향을 알려주는 지표로 현재의 추세를 이용해 이익을 내고자 할 때 사용되는 보조지표입니다. 대표적인 추세지표로는 'MACD'가 있으며, '이동평균선'도 추세 지표에 해당 됩니다.<br>

2. 모멘텀 지표<br>
현재 가격과 일정 기간 이전의 가격 차이 또는 비율을 나타낸 것으로, 현재의 추세가 속도를 더하고 있는지 혹은 줄어들고 있는지를 알려주는 추세의 가속도를 측정하는 지표입니다. 대표적인 모멘텀 지표로는 '스토캐스틱(Stochastic)'과 RSI(Relative Strength Index)가 있으며, 이외에도 이격도, 투자심리선, Price Oscillator, 삼선전환도 등이 있습니다.<br>

3. 변동성 지표<br>
현재 가격의 방향이 아닌 가격의 변동성을 기준으로 시각화 한 지표로서 매매시점을 파악하기에는 다소 어려움이 잇을 수 있지만, 시장 상황을 한눈에 파악할 수 있게 도와주는 지표입니다. 대표적인 변동성 지표로는 유명한 볼린저밴드(Bollinger Band)가 있으며, 이외에도 ATR,Envelope,Keltner Channels 등이 있습니다.<br>

4. 시강강도(거래량) 지표<br>
현재의 추세나 변동성이 얼마나 강한지를 보여주는 지표로서, 거래량을 이용해 분석하는 것이 특징입니다. 대표적인 시장강도 지표로는 거래량 이동평균선, OBV(On Balance Volume)이 있으며, 이외에도 MFI,VO,VR 등이 있습니다.<br>

5. 기타<br>
대부분의 기술적 지표는 위의 네가지 분류에 포함되지만, '일목 균형표'와 같이 하나 이상의 성격을 모두 갖춘 기술적 지표들도 있습니다.<br>




## 기술적 지표의 한계
기술적 지표는 편리하고 유용하지만, 반드시 적중하는 것은 아닙니다. 기술적 지표 하나만을 믿고 거래를 하는 것은 편향적인 관점으로서 지양할 필요가 있으며, 이는 손실을 입게 될 가능성을 크게 키울 수 있습니다. 따라서 여러 가지 기술적 지표를 포함한 기술적 분석과 기본적 분석을 동시에 적용하여 다양한 관점에서 시세를 파악하는 것이 중요합니다.<br>

## 대표적인 기술적 지표

## 1. 이동 평균선(SMA)
이동 평균은 시장 잡음을 걸러내고 매끄러운 가격 움직임을 보여주며, **추세의 방향**을 강조합니다.<br> 

이동 평균은 과거 데이터에 기초하는 후행성 지표입니다.<br>

가장 많이 사용되는 이동 평균은 단순 이동 평균(SMA 또는 MA)과 지수 이동 평균(EMA)입니다.<br>

단순 이동 평균은 특정 기간 동안의 가격 데이터를 참조해 평균을 산출합니다.<br> 예를 들어, 10일 단순 이동 평균은 지난 10일간의 평균 가격을 계산해 표시됩니다.<br> 

반면, 지수 이동 평균은 최근 가격에 더 많은 가중치를 둡니다. 이는 최근의 가격 움직임에 보다 민감하게 반응합니다.<br>

앞서 언급한 것처럼, 이동 평균은 후행성 지표입니다. 기간이 길수록, 더 큰 지체가 발생합니다.<br>

따라서 최근 가격에 대해 200일 단순 이동 평균은 50일 단순 이동 평균보다 느리게 반응할 것입니다.<br>

매수자들은 현 시장 추세를 판단하기 위해 가격과 특정 이동 평균의 상관 관계를 활용하곤 합니다. 예를 들어, 가격이 200일 단순 이동 평균보다 높게 유지되고 있다면, 많은 트레이더들은 해당 자산이 불 마켓(bull market)에 있다고 간주할 수 있습니다.<br>

매수자들은 또한 이동 평균을 매수 또는 매도 신호로 번갈아 가며 사용할 수 있습니다. 예를 들어, 100일 단순 이동 평균이 200일 단순 이동 평균 아래로 교차하게 될 경우 이는 매도 신호일 수 있습니다. 그런데 이러한 교차는 정확히 무엇을 의미하는 것일까요? 이는 지난 100일간의 평균 가격이 지난 200일간의 평균 가격보다 낮다는 것을 의미합니다.<br>

이를 매도 신호로 보는 관점은 단기간의 가격 움직임이 더 이상 상승 추세를 따라가지 않고 있으며, 추세가 뒤바뀔 것이라 보는 것입니다.

## 2. 상대적 강도 지수(RSI)
RSI(Relative Strength Index)
상대강도지수 RSI는 주식, 선물, 옵션 등의 기술적 분석에 사용되는 보조 지표입니다. RSI는 가격의 상승압력과 하락압력 간의 상대적인 강도를 나타냅니다. (1978년 미국의 윌레스 와일더가 개발)  
RSI는 일정 기간 동안 주가가 전일 가격에 비해 상승한 변화량과 하락한 변화량의 평균값을 구하여, 상승한 변화량이 크면 과매수로, 하락한 변화량이 크면 과매도로 판단하는 방식입니다.

> 가격이 전일 가격보다 상승한 날의 상승분은 Up(U)값이라 하고,  
> 가격이 전일 가격보다 하락한 날의 하락분은 Down(D)값이라 합니다.  
> U값과 D값의 평균값을 구하여 그것을 각각 Average Ups(AU), Average Downs(AD)라 합니다.  
> 이 때 AU를 AD값으로 나눈 것을 RS 값이라고 합니다. RS 값이 크다는 것은 일정 기간 동안 하락한 폭보다 상승한 폭이 크다는 것을 의미합니다.  

[RSI 계산 공식]
> RS = AU / AD  
> RSI = RS / (1 + RS) = AU / (AU + AD)  
> 대체로 백분율로 나타냅니다.  

RSI는 기간을 조절 가능한 특징이 있는데, 대체적으로 9일, 14-15일, 25-28일을 사용하고, 개발자인 Wilder는 14일을 권장.  

RSI 그래프는 이동평균선을 함께 나타내는 것이 보통이며, 이동평균선을 며칠선으로 할 것인가 역시 파라미터로 주어집니다. RSI를 15일에 대하여 구하고 5일 평균선을 함께 표시하는 경우 그래프에 (15,5)라고 표시해주는 것이 일반적입니다.  

유사한 지표로는 Stochastic이 있는데, RSI 그래프의 형태는 Fast Stochastic과 비슷하게 나옵니다.

> 지표의 파라미터를 14일로 했을 경우,
* RSI = 14일의 상승폭 합계 / (14일의 상승폭 합계 + 14일의 하락폭 합계)  
* 수식에 따라 RSI는 0과 1 사이의 값.
* 만일 14일 동안 매일 상승하면 '1', 매일 하락하면 '0'이 나옴.
* 보통 0.7을 과열, 0.3을 침체로 봅니다.

### RSI 계산공식
1. U = 전날 주가보다 오늘 주가가 상승할 때의 주가 상승폭(up)  
2. D = 전날 주가보다 오늘 주가가 하락할 때의 주가 하락폭(down)  
3. AU = 일정기간(N일,보통14일) 동안의 U의 평균값(average ups)  
4. AD = 일정기간(N일,보통14일) 동안의 D의 평균값(average downs)  
5. RS = AU / AD  
6. RSI = AU / (AU+AD) = RS / (1+RS)  
7. RSI 시그널 = RSI의 이동평균선 (6일의 이동평균선 사용)

```python
#RSI 구하기
import numpy as np
import pandas as pd
U = np.where(samsung['Close'].diff(1) > 0, samsung['Close'].diff(1), 0) 
D = np.where(samsung['Close'].diff(1) < 0, samsung['Close'].diff(1) *(-1), 0) 
AU = pd.DataFrame(U, index=samsung.index).rolling(window=14).mean() 
AD = pd.DataFrame(D, index=samsung.index).rolling(window=14).mean() 
RSI = AU / (AD+AU) *100 
samsung['RSI'] = RSI
samsung['RSI_signal'] = samsung["RSI"].rolling(6).mean()

#RSI 추가하기
RSI = go.Scatter(x=samsung.index, y=samsung['RSI'], line=dict(color='red', width=2), name='RSI', legendgroup='group5', legendgrouptitle_text='RSI')
```

```python
#과매수매도구간 데이터 추가 (선으로 표현하기 위하여)
samsung['과매수구간'] = 70
samsung['과매도구간'] = 30
#그래프 객체 넣기
fig = make_subplots(specs=[[{"secondary_y": True}]])

# creating first plot
fig.add_trace(
    go.Scatter(x=samsung.index, y=samsung['RSI'], name='삼성전자RSI'),
    secondary_y=False,
)   

# creating next plot
fig.add_trace(
    go.Scatter(x=samsung.index, y=samsung['RSI_signal'], name='삼성전자RSI_signal') , 
   
    secondary_y=True,
)


# creating next plot
fig.add_trace(
    go.Scatter(x=samsung.index, y=samsung['과매수구간'], name='과매수구간') , 
   
    secondary_y=True,
)

# creating next plot
fig.add_trace(
    go.Scatter(x=samsung.index, y=samsung['과매도구간'], name='과매도구간') , 
   
    secondary_y=True,
)
fig.update_layout(title_text="삼성전자 RSI", title_x=0.5)
fig.show()
```

### RSI 해석
1. RSI는 50%를 기준으로, 50% 이상은 매수세 우세, 50% 이하는 매도세 우세를 나타냅니다.
2. RSI가 70% 이상이면 매수세가 상당히 커서 초과매수 국면에 돌입했다고 판단합니다.
3. RSI가 30% 이하이면 매도세가 상당히 커서 초과매도 국면에 돌입했다고 판단합니다.
4. RSI가 시그널선을 상향돌파하면 단기적으로 매수세가 늘어나는 추세라고 판단합니다.
5. RSI가 시그널선을 하향돌파하면 단기적으로 매도세가 늘어나는 추세라고 판단합니다.

## 3. 이동 평균 수렴 확산(MACD)

이동 평균 수렴 확산(MACD)은 특정 자산에서 두 이동 평균 간의 관계를 보여주는 것으로 모멘텀을 결정하는 데 사용됩니다. 이는 이동 평균 수렴 확산선과 신호선, 두 개의 선으로 구성됩니다.<br>

이동 평균 수렴 확산선은 12 지수 이동 평균에서 26 지수 이동 평균을 추출하여 계산됩니다.<br>

이후 이동 평균 수렴 확산선의 신호선인 9 지수 이동 평균 위에 표시됩니다. 많은 차트 분석 도구에는 히스토그램이 포함되곤 하는데, 이는 이동 평균 수렴 확산선과 신호선의 간격을 보여줍니다.<br>

이동 평균 수렴 확산과 가격 움직임 간의 차이를 살펴봄으로써, 트레이더는 현 추세의 강도에 대한 인사이트를 얻을 수 있습니다.<br>

예를 들어, 가격이 더 높은 고점을 형성했으나 이동 평균 수렴 확산 고점이 낮아졌을 경우, 머잖아 시장 추세가 반전될 수 있습니다. 그렇다면 이 경우 이동 평균 수렴 확산을 통해 무엇을 알 수 있을까요? 모멘텀이 약해지는 동안 가격이 상승한다면, 가격이 하락하거나 추세가 역전될 확률이 높다는 것을 의미합니다.<br>

투자자는 이동 평균 수렴 확산선과 신호선이 교차하는 순간에 이를 활용할 수도 있습니다.<br>

예를 들어, 이동 평균 수렴 확산선이 신호 선 위로 교차할 경우, 이는 매수 신호일 수도 있습니다. 역으로, 이동 평균 수렴 확산선이 신호선 아래로 내려간다면, 매도 신호일 수도 있습니다.<br>

이동 평균 수렴 확산은 보통 상대적 강도 지수(RSI)와 함께 사용되는데, 두 가지 지표가 서로 다른 요소를 통해 모멘텀을 측정하기 때문입니다.<br>

이는 두 가지를 함께 사용하면 시장에 대한 보다 온전한 기술적 전망을 얻을 수 있다는 가정에 기반하고 있습니다.

### MACD의 세가지 지표
> 1. MACD = 12일 이동평균선 - 26일 이동평균선  
> 2. MACD Signal = MACD의 9일 이동평균선  
> 3. 0선 = 지표값의 양/음을 나타내는 기준선

### MACD Oscillator

## 4. 스토캐스틱(stochastic)

특정 시장의 추세를 측정하는 데 사용되는 지표입니다. 스토캐스틱은 당일의 종가 위치를 백분율로 표시하며, 최근 N일간의 최고가와 최저가의 범위 내에서 현재 가격의 위치를 표시할 때, 매수세가 매도세보다 강할 때는 그 위치가 높게 형성되고 매도세가 매수세보다 강할 때는 그 위치가 낮게 형성된다는 것을 나타냅니다.<br>

스토캐스틱 값은 (현재 가격 - N일중 최저가)/(N일중 최고가 - N일중 최저가)로 계산되며 보통 백분율로 표기합니다. 스토캐스틱 값의 범위는 항상 0~100% 사이가 되는데요. 100%라면 현재 가격이 N일간 최고가이므로 매수세가 가장 강한 경우가 되며, 0%라면 현재 가격이 N일간 최저가이므로 매도세가 가장 강한 경우가 됩니다.<br>



만약 N을 15로 하면, 스토캐스틱은 15일중 최고가와 최저가를 이용하는 값이 됩니다. 예를 들어 최근 15일 중 최고가가 15,000원이고 최저가가 10,000원이며 현재가격이 14,000원이라면 스토캐스틱 값은 80%가 됩니다. 고가가 15,000원이고 저가가 10,000원이며 현재가격이 11,000원이라면 스토캐스틱 값은 20%가 되죠.<br>



스토캐스틱이 일반적으로 80이 넘으면 과매수 상태인 것으로 간주되며 20 미만은 과매도 상태로 간주됩니다. 이는 해당 주식이 범위의 최고점 주변이나 최저점 주변에서 거래되고 있음을 의미합니다.<br>



스토캐스틱 차트에는 두가지 그래프가 표시됩니다. 하나는 스토캐스틱이고 두번째는 스토캐스틱의 N일 이동평균선입니다. 전자를 K로, 후자를 D로 표시하는데요. D는 단순히 스토캐스틱의 이동평균선이므로 slow K라고 부르기도 하죠. 이동평균선을 함께 표시하는 이유는 골든크로스와 데드크로스 등 이동평균선과의 비교를 통해 스토캐스틱 값의 변곡점을 쉽게 파악할 수 있기 위해서입니다.<br>


K와 D를 표시한 스토캐스틱 차트에서, K가 증가할 경우 즉 매수세가 증가하면 이동평균선인 D를 뚫고 올라오는 형태 즉 골든크로스를 보이게 되는데 이것은 매수 신호로 볼 수 있습니다. 반대로 K가 D를 뚫고 내려오는 형태 즉 데드크로스를 보이면 매도 신호로 읽을 수 있습니다.<br>


```python
# 스토캐스틱 구하기 N일을 14일로 설정하겠습니다.
samsung['ndays_high'] = samsung['High'].rolling(window=14, min_periods=1).max() # 14일 중 최고가 
samsung['ndays_low'] = samsung['Low'].rolling(window=14, min_periods=1).min() # 14일 중 최저가 
samsung['fast_k'] = (samsung['Close'] - samsung['ndays_low']) / (samsung['ndays_high'] - samsung['ndays_low']) * 100 # Fast %K 구하기 
samsung['slow_d'] = samsung['fast_k'].rolling(window=3).mean() # Slow %D 구하기
```

```python
# 스토캐스틱 지표 생성하기
fast_k = go.Scatter(x=samsung.index, y=samsung['fast_k'], line=dict(color='green', width=2), name='fast_k') 
slow_d = go.Scatter(x=samsung.index, y=samsung['slow_d'], line=dict(dash='dashdot', color='black', width=2), name='slow_d') 
```

```python
#그래프 객체 넣기
fig = ms.make_subplots(rows=3, cols=1,  shared_xaxes=True,shared_yaxes=False, horizontal_spacing=0.05, vertical_spacing=0.02) 

fig.add_trace(candle,row=1,col=1) 
fig.add_trace(fast_k,row=3,col=1) 
fig.add_trace(slow_d,row=3,col=1) 
fig.update_layout(title_text="삼성전자 Stochastic", title_x=0.5)
fig.show()
```

## 5. 볼린저 밴드(Bollinger Bands)
볼린저 밴드는 시장 변동성을 측정하는 동시에, 과매수와 과매도 상태를 파악하는데 사용됩니다.<br>

볼린저 밴드는 단순 이동 평균(중간 선), 위쪽 선, 아래쪽 선, 세 개의 선으로 구성됩니다. 설정은 다양할 수 있지만, 위쪽과 아래쪽 선은 일반적으로 중간 선의 표준 편차입니다. 변동성이 증가하거나 감소하면, 밴드의 폭 또한 커지거나 줄어듭니다.<br>

보통 가격이 위쪽 선에 가까워질수록 차트상의 자산은 과매수 상태에 가깝다는 것을 의미합니다. 역으로, 가격이 아래쪽 선에가까워질수록 과매도 상태일 수 있음을 뜻합니다.<br>

대부분의 경우 가격은 밴드 안에 있기 마련이지만, 드물게 아래나 위를 돌파하기도 합니다. 이러한 경우는 그 자체로 트레이딩 신호는 아닐 수 있지만, 극단적인 시장 상황을 보여주는 것일 수 있습니다.<br>

볼린저 밴드에서 중요한 또 한 가지는 스퀴즈(squeeze)라 하는 것입니다. 이는 변동성이 낮은 기간을 가리키는데, 모든 선이 아주 가깝게 붙어 있습니다. 이는 미래의 잠재적인 변동성을 암시하는 것일 수도 있습니다. 이와 반대로, 선들이 몹시 멀리 떨어져 있다면, 변동성이 줄어드는 기간이 찾아올 수도 있습니다.<br>


### 볼린저 밴드의 구성요소
> * N기간 동안의 이동평균(MA)
* 이동평균 위의 K배 표준편차 상위 밴드
* 이동평균 아래의 K배 표준편차 하위 밴드

일반적으로 N과 K의 값은 20과 2입니다. 평균값의 기본 선택의 단순 이동 평균이지만, 필요에 따라 다른 종류의 평균값을 쓸 수 있습니다. 그래도, 중간 밴드와 표준 편차의 계산을 위해 같은 기간을 사용합니다.

```python
# 볼린저밴드 구하기 
samsung['ma20'] = samsung['Close'].rolling(window=20).mean() # 20일 이동평균 
samsung['stddev'] = samsung['Close'].rolling(window=20).std() # 20일 이동표준편차 
samsung['upper'] = samsung['ma20'] + 2*samsung['stddev'] # 상단밴드 
samsung['lower'] = samsung['ma20'] - 2*samsung['stddev'] # 하단밴드
```

```python
# 볼린저밴드 상단/하단생성하기
upper = go.Scatter(x=samsung.index, y=samsung['upper'], line=dict(color='red', width=2), name='bol_upper') 
lower = go.Scatter(x=samsung.index, y=samsung['lower'], line=dict(color='blue', width=2), name='bol_lower') 
```

```python
#그래프 객체 넣기
fig = go.Figure(data=[candle,upper,lower,ma5, ma20, ma60])
fig.update_layout(title_text="삼성전자 Bollinger Bands", title_x=0.5)
fig.show()
```

```python
# ! pip install backtrader #백트레이더 라이브러리 설치
```

```python
import backtrader as bt
from pykrx import stock
import pandas as pd
from datetime import datetime
from datetime import timedelta
from backtrader import cerebro
from pandas.io.formats import style
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

class SmaCross(bt.Strategy): # bt.Strategy를 상속한 class로 생성해야 함. (이동평균선)

    params = dict(
        pfast=5,  # period for the fast moving average   5평선
        pslow=20   # period for the slow moving average  20평선
    )

    def log(self, txt, dt=None): # 매수/매도 기록 함수
        ''' Logging function fot this strategy'''
        dt = dt or self.data.datetime[0]
        if isinstance(dt, float):
                dt = bt.num2date(dt)
                print('%s, %s' % (dt.isoformat(), txt))

    def __init__(self):
        sma1 = bt.ind.SMA(period=self.p.pfast)  # fast moving average
        sma2 = bt.ind.SMA(period=self.p.pslow)  # slow moving average
        self.crossover = bt.ind.CrossOver(sma1, sma2)  # crossover signal


  
    def next(self):  #전략 함수
        global size
        if not self.position: # not in the market # 포지션이 없으면(매수/매도가 없음)
            if self.crossover > 0:  # 단기 이동 평균선이 장기 이동 평균선을 상향 돌파시 매수
              
                size = int(self.broker.getcash() / self.data.close[0])
                # 최대 구매 가능 개수
                self.buy(price=self.data.close[0], size=size) #매수
                # 매수 size = 구매 개수 설정
                self.log('BUY CREATE, %.2f' % (self.data.close[0]))
                print(size, 'EA')
        else: #포지션이 있는 경우
            if self.crossover < 0:  # 단기 이동 평균선이 장기 이동 평균선을 하향 돌파시 매도
                self.sell(price=self.data.close[0],size=size)
                # 매도
                self.log('SELL CREATE, %.2f' % (self.data.close[0]))
                print(size, 'EA')
#--------------------------------------------------------------------

# cerebro 가져오기
cerebro = bt.Cerebro()

ticker =  "005930"   # 삼성전자 종목코드
# 전체 종목코드와 종목명 가져오기
# df = df.rename(columns={'시가':'open', '고가':'high', '저가':'low', '종가':'close', '거래량':'volume'})
# df["open"]=df["open"].apply(pd.to_numeric,errors="coerce")

from pykrx import stock
df = stock.get_market_ohlcv("2021-12-01", "2022-07-01", ticker)
df.columns = ["open","high","low","close","volume"]

data = bt.feeds.PandasData(dataname=df) #데이터 넣어주기
# 데이터 추가
cerebro.adddata(data)

#cerebro 실행
cerebro.addstrategy(SmaCross)  # 투자 전략 추가
cerebro.broker.setcash(10000000)  # 브로거 설정

# 매매 단위 설정하기
cerebro.addsizer(bt.sizers.SizerFix, stake=100) 

# 초기 투자금
init_cash = cerebro.broker.getvalue()
cerebro.run()

# 최종 금액
final_cash = cerebro.broker.getvalue()

print("최종 금액 : ", final_cash, "\\")
print("수익률 : ", float(final_cash - init_cash) / float(init_cash) *100., "%")

cerebro.plot()
```

매수 매도는 정상적으로 되는것으로 확인됩니다.<br>

Backtrader가 코랩에서 그래프가 즉각적으로 표현되지 않고있어서 다른 백테스팅 라이브러리인 Backtesting을 설치하여 같은 전략으로 그래프를 띄워 보았습니다.<br>

```python
# ! pip install Backtesting #백테스팅 설치
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA



class SmaCross(Strategy):
    def init(self):
        Close = self.data.Close
        self.ma1 = self.I(SMA, Close, 5) #5평선
        self.ma2 = self.I(SMA, Close, 20) #20평선

    def next(self):
        if crossover(self.ma1, self.ma2): #골든크로스
            self.buy()
        elif crossover(self.ma2, self.ma1): #데드크로스
            self.sell()

ticker =  "005930"   # 삼성전자 종목코드
# 전체 종목코드와 종목명 가져오기
# df = df.rename(columns={'시가':'open', '고가':'high', '저가':'low', '종가':'close', '거래량':'volume'})
# df["open"]=df["open"].apply(pd.to_numeric,errors="coerce")

from pykrx import stock
df = stock.get_market_ohlcv("2021-12-01", "2022-07-01", ticker)
df.columns = ["Open","High","Low","Close","Volume"]

# 초기투자금 10000, commission 비율 0.002 임의 지정
bt = Backtest(df, SmaCross,
              cash=10000, commission=.002)

bt.run()
bt.plot()
```

5평선과 20평선이 백테스트를 통해 나타나면서 크로스 오버가나는 날마다 매수 매도가 이루어 지고 있습니다.<br>

마지막 매수날짜를 기점으로 미세하게 데드크로스가 발생하면서 주가가 급격하게 떨어졌습니다만 매도가 이루어 지지 않은 것으로 보아 골든크로스와 데드크로스사이가 너무 빨라서 그런 것인지, 세부적인 옵션을 통해 정확하게 백테스팅을 시도할 필요가 있어 보입니다.<br>
