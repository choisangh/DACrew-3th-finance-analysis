ECOSÏóêÏÑú Ï†úÍ≥µÌïòÎäî Í≤ΩÍ∏∞ ÏÑ†Ìñâ ÏßÄÏàòÏôÄ ÎπÑÍµêÌïòÏó¨ ÏÇ¥Ìé¥Î≥¥Í≤†ÏäµÎãàÎã§.<br>
Í≤ΩÍ∏∞ ÏÑ†Ìñâ ÏßÄÏàòÎäî Ïã§Ï†ú Í≤ΩÍ∏∞ÏàúÌôòÏóê ÏïûÏÑú Î≥ÄÎèôÌïòÎäî Í∞úÎ≥ÑÏßÄÌëúÎ•º Í∞ÄÍ≥µ„ÜçÏ¢ÖÌï©ÌïòÏó¨ ÎßåÎì† ÏßÄÏàòÎ°ú Ìñ•ÌõÑ Í≤ΩÍ∏∞Î≥ÄÎèôÏùò Îã®Í∏∞ ÏòàÏ∏°Ïóê Ïù¥Ïö©Îê©ÎãàÎã§. <br>


```python
# ECOS Îç∞Ïù¥ÌÑ∞ Ìò∏Ï∂ú Ìï®Ïàò
def EcosDownload(Statcode, Freq, Begdate, Enddate, Subcode1, Subcode2, Subcode3):
    
    url = 'http://ecos.bok.or.kr/api/StatisticSearch/HA5OFFNUUGG4TD13CMHV/xml/kr/1/100000/%s/%s/%s/%s/%s/%s/%s/'%(Statcode, Freq, Begdate, Enddate, Subcode1, Subcode2, Subcode3)

    raw = requests.get(url)
    xml = BeautifulSoup(raw.text,'xml')

    raw_data = xml.find_all("row")

    date_list = []
    value_list = []

    for item in raw_data:
         
            value = item.find('DATA_VALUE').text.encode('utf-8')
            date_str = item.find('TIME').text
            
            value = float(value)
            try:
                date_list.append(datetime.datetime.strptime(date_str,'%Y%m'))
            except:
                date_list.append(datetime.datetime.strptime(date_str,'%Y%m%d'))
            value_list.append(value)
            
    df = pd.DataFrame(index = date_list)
    
    df['value'] = value_list
      
    return df
```

```python
kospi = stock.get_index_ohlcv("20000101", "20220531", "1001") #20ÎÖÑÏπò ÏΩîÏä§Ìîº ÏßÄÏàò ohlcv
```

```python
LEI=EcosDownload('901Y067', 'M', '200001', '202207', 'I16E', '', '') #Í≤ΩÍ∏∞ ÏÑ†ÌñâÏßÄÏàò ÏàúÌôòÎ≥ÄÎèôÏπò Îç∞Ïù¥ÌÑ∞ load
```

```python
#ÏãúÍ∞ÑÎã®ÏúÑÎ•º ÎßûÏ∂îÍ∏∞ ÏúÑÌï¥ ÏõîÎã®ÏúÑÎ°ú resampling
LEI=LEI.resample('1M').last()
kospi=kospi.resample('1M').last()
```

```python
df=pd.merge(LEI,kospi,left_index=True,right_index=True,how='outer')
```

```python
plt.rcParams['figure.figsize'] = (12, 6)

fig, ax1 = plt.subplots()
ax1.plot(df.index, df['Ï¢ÖÍ∞Ä'], color='blue')

ax2 = ax1.twinx()
ax2.plot(df.index, df['value'], color='red')

plt.show()
```

## ÏàúÌôòÎ≥ÄÎèôÏπò ÏÇ∞Ï∂ú (DOUBLE HP FILTER)
* ref : https://towardsdatascience.com/introduction-to-trend-filtering-with-applications-in-python-d69a58d23b2

ÏÇ∞Ï∂úÌïú ÏßÄÏàòÏùò Î∂àÍ∑úÏπôÏÑ±ÏùÑ Ï†úÍ±∞ÌïòÍ≥† ÌÅ∞ Ï∂îÏÑ∏Î•º ÌååÏïÖÌïòÍ∏∞ ÏúÑÌï¥ÏÑú HP FILTERÎ•º ÌôúÏö©ÌïòÏó¨ ÏàúÌôòÎ≥ÄÎèôÏπòÎ•º ÏÇ∞Ï∂úÌï©ÎãàÎã§.

```python
import pandas as pd
import numpy as np
import cvxpy 
import scipy
import cvxopt 
import matplotlib.pyplot as plt
```

```python
y = final['ÏßÄÏàò'].to_numpy()
y = np.log(y)
n = y.size
ones_row = np.ones((1, n))
D = scipy.sparse.spdiags(np.vstack((ones_row, -2*ones_row, ones_row)), range(3), n-2, n)
```

```python
lambda_list = [0, 0.1, 0.5, 1, 2, 5, 10, 50, 200, 500, 1000, 2000, 5000, 10000, 100000]
```

lambdaÍ∞íÏùÑ ÌÜµÌï¥ÏÑú Ï∂îÏÑ∏Ïóê ÎåÄÌïú ÎîîÌÖåÏùºÌïú Ï°∞Ï†ïÏùÑ Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
solver = cvxpy.CVXOPT
reg_norm = 2
fig, ax = plt.subplots(len(lambda_list)//3, 3, figsize=(20,20))
ax = ax.ravel()
ii = 0
for lambda_value in lambda_list:
    x = cvxpy.Variable(shape=n) 
    # x is the filtered trend that we initialize
    objective = cvxpy.Minimize(0.5 * cvxpy.sum_squares(y-x) 
                  + lambda_value * cvxpy.norm(D@x, reg_norm))
    # Note: D@x is syntax for matrix multiplication
    problem = cvxpy.Problem(objective)
    problem.solve(solver=solver, verbose=False)
    ax[ii].plot(np.arange(1, n+1), y, linewidth=1.0, c='b')
    ax[ii].plot(np.arange(1, n+1), np.array(x.value), 'b-', linewidth=1.0, c='r')
    ax[ii].set_xlabel('Time')
    ax[ii].set_ylabel('Log Premium')
    ii+=1
    
plt.tight_layout()
plt.show()
```

```python
x = cvxpy.Variable(shape=n) 
objective = cvxpy.Minimize(0.5 * cvxpy.sum_squares(y-x) 
                + 0.5 * cvxpy.norm(D@x, reg_norm))
problem = cvxpy.Problem(objective)
problem.solve(solver=solver, verbose=False)
np.array(x.value)
```
